```python
# Plot a function and its integral
from math import *
import matplotlib.pyplot as plt
fctn = input("Enter a function of x: ")
a = float(input("Enter starting value of x: "))
b = float(input("Enter ending value of x: "))
N = int( input("Enter number of samples: "))
def f(x):
    return eval(fctn)
# Initialize and plot first point
I = 0.
x = a
dx = (b-a)/(N-1)
d = (f(a + dx) - f(a))/dx
y = f(a)
plt.plot(x, y, 'k-', label=r"$f(x)$")
plt.plot(x, I, 'r-', label=r"$\int_0^x f(x')\,dx'$")
plt.plot(x, d, 'b-', label=r"dy/dx")
while x < b:
    # step forward and calculate integral
    xn = x + dx
    yn = f(xn)
    print(f"x: {xn} \t y: {yn}")
    In = I + 0.5*(y+yn)*dx # trapezoidal rule
    # step forward and calculate derivative
    dn = (f(xn) - f(x))/dx
    # plot line connecting old and new points
    plt.plot([x, xn], [y, yn], 'k-')
    plt.plot([x, xn], [I, In], 'r-')
    plt.plot([x, xn], [d, dn], 'b-')
    # update; new point becomes old point
    x = xn
    y = yn
    I = In
    d = dn
plt.xlabel(r'$x$')
plt.ylabel('Function and integral')
plt.legend()
plt.show()
print("final value of integral = ", I)
```
